```python
# Reading CSV of Dataframe
Event_Information_Dataframe1 = pd.read_csv('/Users/devonmarcus/Data Science Bootcamp/Ticketmaster Project/data/Event_Information_Dataframe1.csv', sep='\t')
Event_Information_Dataframe2 = pd.read_csv('/Users/devonmarcus/Data Science Bootcamp/Ticketmaster Project/data/Event_Information_Dataframe2.csv', sep='\t')
Event_Information_Dataframe3 = pd.read_csv('/Users/devonmarcus/Data Science Bootcamp/Ticketmaster Project/data/Event_Information_Dataframe3.csv', sep='\t')
Event_Information_Dataframe4 = pd.read_csv('/Users/devonmarcus/Data Science Bootcamp/Ticketmaster Project/data/Event_Information_Dataframe4.csv', sep='\t')
Event_Information_Dataframe5 = pd.read_csv('/Users/devonmarcus/Data Science Bootcamp/Ticketmaster Project/data/Event_Information_Dataframe5.csv', sep='\t')
```

```python
# Concatenating Dataframes Together
df = [Event_Information_Dataframe1, Event_Information_Dataframe2, Event_Information_Dataframe3,Event_Information_Dataframe4,Event_Information_Dataframe5]
df = pd.concat(df)
pd.set_option('display.max_columns', 999)
df
```

```python
# Dropping First Column
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
# Dropping Row with Same ID
df = df.drop_duplicates(subset=['ID'], keep='first')
```

```python
# Convert Date Columns to Int
df['Year'] = df['Year'].astype("Int64", errors="ignore")
df['Month'] = df['Month'].astype("Int64", errors="ignore")
df['Day'] = df['Day'].astype("Int64", errors="ignore")
```

```python
# Removing Row with NaN in Essential Columns
df = df.replace(to_replace='"Undefined"', value='"nan"')

df = df[df['Year'].notna()]
df = df[df['Month'].notna()]
df = df[df['Day'].notna()]
df = df[df['Price_Ranges_Min'].notna()]
df = df[df['Price_Ranges_Max'].notna()]
df = df[df['Home_Name'].notna()]
df = df[df['Segment_Home'].notna()]
df = df[df['Event_Tickets'].notna()]
df = df[df['Venue_Country'].notna()]
df = df[df['Price_Ranges_Currency'].notna()]
```

```python
# Creating columns for Date and Day of Week
df['Date'] = pd.to_datetime( df[['Year', 'Month', 'Day']])
df['Day_of_Week'] = df['Date'].dt.day_name()
```

```python
# Converting all Currencies to CAD 
c1 = (df['Price_Ranges_Currency']=='"USD"')
c2 = (df['Price_Ranges_Currency']=='"CAD"')
c3 = (df['Price_Ranges_Currency']=='"AUD"')
c4 = (df['Price_Ranges_Currency']=='"GBP"')
c5 = (df['Price_Ranges_Currency']=='"NZD"')
c6 = (df['Price_Ranges_Currency']=='"EUR"')
r1 = (df['Price_Ranges_Min'] * 1.28)
r2 = (df['Price_Ranges_Min'] * 1)
r3 = (df['Price_Ranges_Min'] * 0.9)
r4 = (df['Price_Ranges_Min'] * 1.56)
r5 = (df['Price_Ranges_Min'] * 0.8)
r6 = (df['Price_Ranges_Min'] * 1.35)
conditions = [c1,c2,c3,c4,c5,c6]
results = [r1,r2,r3,r4,r5,r6]
df['Price_Ranges_Min'] = np.select(conditions,results)

c1 = (df['Price_Ranges_Currency']=='"USD"')
c2 = (df['Price_Ranges_Currency']=='"CAD"')
c3 = (df['Price_Ranges_Currency']=='"AUD"')
c4 = (df['Price_Ranges_Currency']=='"GBP"')
c5 = (df['Price_Ranges_Currency']=='"NZD"')
c6 = (df['Price_Ranges_Currency']=='"EUR"')
r1 = (df['Price_Ranges_Max'] * 1.28)
r2 = (df['Price_Ranges_Max'] * 1)
r3 = (df['Price_Ranges_Max'] * 0.9)
r4 = (df['Price_Ranges_Max'] * 1.56)
r5 = (df['Price_Ranges_Max'] * 0.8)
r6 = (df['Price_Ranges_Max'] * 1.35)
conditions = [c1,c2,c3,c4,c5,c6]
results = [r1,r2,r3,r4,r5,r6]
df['Price_Ranges_Max'] = np.select(conditions,results)
```
